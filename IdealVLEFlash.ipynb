```python
p=Props(['Benzene','Toluene'])
```

```python
def bubbleP(x, T):
    yP = x*p.Pvap(T)
    P = np.sum(yP)
    return P, yP/P
```

```python
def dewP(y, T):
    Pvap = p.Pvap(T)
    P = 1/np.sum(y/Pvap)
    return P, y*P/Pvap
```

```python
def bubbleT(x, P):

    T0 = np.sum(x*p.Tb(P))
    def froot(T):
        return bubbleP(x,T)[0]-P

    return sp.optimize.root_scalar(froot, x0=T0, method='secant').root
```

```python
def dewT(y, P):

    T0 = np.sum(y*p.Tb(P))
    def froot(T):
        return dewP(y,T)[0]-P

    return sp.optimize.root_scalar(froot, x0=T0, method='secant').root
```
