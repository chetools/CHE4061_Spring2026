```python
p=Props(['Benzene','Toluene'])
```

```python
def bubbleP(x, T):
    yP = x*p.Pvap(T)
    P = np.sum(yP)
    return P, yP/P
```

```python
def dewP(y, T):
    Pvap = p.Pvap(T)
    P = 1/np.sum(y/Pvap)
    return P, y*P/Pvap
```

```python
def bubbleT(x, P):

    T0 = np.sum(x*p.Tb(P))
    def froot(T):
        return bubbleP(x,T)[0]-P

    T=sp.optimize.root_scalar(froot, x0=T0, method='secant').root
    return T, x*p.Pvap(T)/P
```

```python
def dewT(y, P):

    T0 = np.sum(y*p.Tb(P))
    def froot(T):
        return dewP(y,T)[0]-P

    T=sp.optimize.root_scalar(froot, x0=T0, method='secant').root
    return T, y*P/p.Pvap(T)
```

```python
#Const T, binary-ideal VLE
T= 360.
x1s = np.linspace(0,1,21)
Ps=[]
y1s=[]
for x1 in x1s:
    P, y = bubbleP(np.array([x1, 1-x1]), T)
    Ps.append(P)
    y1s.append(y[0])

fig = make_subplots()
fig.add_scatter(x=x1s, y=Ps, name='bubbleP')
fig.add_scatter(x=y1s, y=Ps, name='dewP')
fig.update_layout(width=600, height=400)
```

```python
#Const P, binary-ideal VLE
P= 1e5
x1s = np.linspace(0,1,21)
Ts=[]
y1s=[]
for x1 in x1s:
    T, y = bubbleT(np.array([x1, 1-x1]), P)
    Ts.append(T)
    y1s.append(y[0])

fig = make_subplots()
fig.add_scatter(x=x1s, y=Ts, name='bubbleT')
fig.add_scatter(x=y1s, y=Ts, name='dewT')
fig.add_scatter(x=x1s, y=y1s, row=1,col=2)
fig.update_layout(width=600, height=400)
```
