```python
import numpy as np
from plotly.subplots import make_subplots
```

```python
rawTP = np.genfromtxt("WaterVaporPressure.csv", delimiter=",")
T = rawTP[:,0]
logP = np.log10(rawTP[:,1])
```

```python
A = np.c_[T, -logP, np.ones_like(T)]
y = T*logP
```

```python
a, c, d = np.linalg.solve(A.T@A, A.T@y)
b = a*c - d
```

```python
Tplot = np.linspace(T.min(), T.max(), 100)
antoineP = 10**(a-b/(Tplot+c))
```

```python
fig = make_subplots()
fig.add_scatter(x=T, y=10**logP, mode='markers')
fig.add_scatter(x=Tplot, y=antoineP, mode='lines')
fig.update_yaxes(type="log")
```
